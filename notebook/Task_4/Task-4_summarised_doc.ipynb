# E-Commerce Advanced Analytics & Machine Learning (Task 4)

This repository contains the complete implementation for **Task 4**, focusing on Statistical Analysis, Customer Segmentation, and Predictive Modeling on the Online Retail Dataset.

---

##  Project Overview

1. **Statistical Hypothesis Testing (`scipy.stats`):**
   - **Two-Sample t-Test:** Evaluated regional revenue differences between UK and Non-UK markets.
   - **Chi-Square Test of Independence:** Analyzed relationship between order volumes and geographic markets.
   - **95% Confidence Intervals:** Estimated baseline mean revenue using Parametric and Bootstrap methods.

2. **Customer Segmentation (K-Means Clustering):**
   - Feature engineering using **RFM Analysis (Recency, Frequency, Monetary)**.
   - Scaled metrics using `StandardScaler` after log transformation.
   - Grouped customers into **4 Optimal Clusters** and visualized them in 2D space using **PCA**.

3. **Predictive Modeling (Customer Churn Prediction):**
   - Formulated binary target variable `Is_Churn` (> 90 days inactivity).
   - Built and evaluated a **Logistic Regression** model on an 80/20 train/test split.
   - Achieved **82% Classification Accuracy**.
