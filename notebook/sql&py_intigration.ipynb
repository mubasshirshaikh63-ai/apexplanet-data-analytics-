```python
import pandas as pd
from sqlalchemy import create_engine

class DBHelper:
    def __init__(self, db_path):
        self.engine = create_engine(f"sqlite:///{db_path}")

    def run_query(self, query):
        return pd.read_sql(query, self.engine)

# Initialize with your database path
db = DBHelper(r"C:\Users\SKHAKH MUBASHSHIR\Documents\organization.db")
```

```python
# 1. Top 5 products by sales
q1 = """
SELECT City, COUNT(*) AS EmployeeCount
FROM employees
GROUP BY City
ORDER BY EmployeeCount DESC
LIMIT 5;
"""
print(db.run_query(q1))
```

```python
# Distribution of employees by education level
q2 = """
SELECT Education, COUNT(*) AS Count
FROM employees
GROUP BY Education;
"""
print(db.run_query(q2))
```

```python
q3 = """
SELECT City, AVG(Age) AS AvgAge
FROM employees
GROUP BY City;
"""
print(db.run_query(q3))
```

```python
#
q4 = """
SELECT Gender, COUNT(*) AS Count
FROM employees
GROUP BY Gender;
"""
print(db.run_query(q4))
```

```python
#Payment tier distribution
q5 = """
SELECT PaymentTier, COUNT(*) AS Count
FROM employees
GROUP BY PaymentTier;
"""
print(db.run_query(q5))
```

```python
#Employees who ever got benched
q6 = """
SELECT EverBenched, COUNT(*) AS Count
FROM employees
GROUP BY EverBenched;
"""
print(db.run_query(q6))
```

```python
#Average experience in current domain by education
q7 = """
SELECT Education, AVG(ExperienceInCurrentDomain) AS AvgExperience
FROM employees
GROUP BY Education;
"""
print(db.run_query(q7))
```

```python
#Attrition analysis (LeaveOrNot) by city
q8 = """
SELECT City, SUM(LeaveOrNot) AS EmployeesLeft, COUNT(*) AS TotalEmployees
FROM employees
GROUP BY City;
"""
print(db.run_query(q8))
```

```python
#Attrition by gender
q9 = """
SELECT Gender, SUM(LeaveOrNot) AS EmployeesLeft, COUNT(*) AS TotalEmployees
FROM employees
GROUP BY Gender;
"""
print(db.run_query(q9))
```

```python
#Attrition by payment tier
q10 = """
SELECT PaymentTier, SUM(LeaveOrNot) AS EmployeesLeft, COUNT(*) AS TotalEmployees
FROM employees
GROUP BY PaymentTier;
"""
print(db.run_query(q10))
```

```python
#Reusable Utility Script
import pandas as pd
from sqlalchemy import create_engine

class DBHelper:
    def __init__(self, db_path):
        self.engine = create_engine(f"sqlite:///{db_path}")

    def run_query(self, query):
        return pd.read_sql(query, self.engine)

# Usage Example
db = DBHelper(r"C:\Users\SKHAKH MUBASHSHIR\Documents\organization.db")

# Run one of the queries
df = db.run_query("""
SELECT City, COUNT(*) AS EmployeeCount
FROM employees
GROUP BY City
ORDER BY EmployeeCount DESC
LIMIT 5;
""")
print(df)
```
